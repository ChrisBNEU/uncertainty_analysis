```python
key= 'Surface_Abstraction/Abstracting;Donating/alpha'
sobol_range_map[key][0]
sobol_map[key][1]
```

```python
for key, value in sobol_map.items():
    
    y_array = len(sobol_map[key][1])*[1]
    
    if key.endswith("/A") and sobol_range_map[key][0] > 1:
        mid = 10**sobol_range_map[key][0]
        minn = 10**sobol_range_map[key][1]
        maxx = 10**sobol_range_map[key][2]
        
        plt.scatter(mid, 1)
        plt.scatter([minn, maxx], [1,1])
        plt.scatter(sobol_map[key][1], y_array, s = 10)
        
        plt.xscale('log')
        
    elif key.endswith("/E0") or key.endswith("Ea"):
        ea2ev = 9.6e4                              
        mid = sobol_range_map[key][0]/ea2ev
        minn = sobol_range_map[key][1]/ea2ev
        maxx = sobol_range_map[key][2]/ea2ev
        
        plt.scatter(mid, 1)
        plt.scatter([minn, maxx], [1,1])
        plt.scatter(sobol_map[key][1]/ea2ev, y_array, s = 10)

                                             
    else: 
        plt.scatter(sobol_range_map[key][0], 1)
        plt.scatter(sobol_range_map[key][1:], [1,1])
        plt.scatter(sobol_map[key][1], y_array, s = 10)
          
          
          
    plt.suptitle(key)
        

        
        
    plt.show()
    
    print(sobol_range_map[key])
```
